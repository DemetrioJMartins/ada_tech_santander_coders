# **Perguntas e respostas sobre os dados**

## ***1 - Quais são os 10 filmes com maior avaliação ?***

```python
query = """
SELECT
  originalTitle AS titulo,
  averageRating As nota,
  directors
FROM imdb
ORDER BY nota DESC
LIMIT 10
"""

result = sqldf(query, locals())
print(result)
```

## ***2 - Os filmes mais mal avaliados***

```python
query = """
SELECT
  originalTitle AS titulo,
  averageRating AS nota,
  directors
FROM imdb
ORDER BY averageRating ASC
LIMIT 10
"""

result = sqldf(query, locals())
print(result)
```

## ***3 - Melhores filmes com menos de 120 minutos***

```python
query = """
SELECT
  originalTitle AS titulo,
  runtimeMinutes AS minutos_de_filme,
  averageRating as nota,
  directors
FROM imdb
WHERE runtimeMinutes < 90
ORDER BY averageRating DESC
LIMIT 10
"""

result = sqldf(query, locals())
print(result)
```

## ***4. Avaliação dos 15 filmes mais caros***

```python
query = """
SELECT
  originalTitle AS titulo,
  averageRating AS nota,
  budget AS custo,
  directors
FROM imdb
ORDER BY budget DESC
LIMIT 15
"""

result = sqldf(query, locals())
print(result)
```

## ***5 - Filmes que tiveram o maior 'lucro' (O custo de produção não engloba todo os custos, como os repasses dos cinemas)***

```python
query = """
SELECT
  originalTitle AS titulo,
  budget AS custo,
  gross AS bilheteria,
  profit AS lucro,
  directors
FROM imdb
ORDER BY profit DESC
LIMIT 15
"""

result = sqldf(query, locals())
print(result)
```

## ***6 - Melhores filmes que custaram mais de 100 milhões***

```python
query = """
SELECT
  originalTitle AS titulo,
  averageRating AS nota,
  budget
FROM imdb
WHERE budget > 100000000
ORDER BY averageRating DESC
LIMIT 10
"""

result = sqldf(query, locals())
print(result)
```

## ***7. Filmes de 2023 com custo abaixo de 100 milhões que mais lucraram***

```python
query = """
SELECT
  originalTitle AS titulo,
  release_date AS data_de_lancamento,
  averageRating AS nota,
  budget
FROM imdb
WHERE release_date LIKE '%2023' AND budget < 100000000
ORDER BY averageRating DESC
LIMIT 10
"""

result = sqldf(query, locals())
print(result)
```

## ***8 - Diretores possuem a maior média de avaliação***



```python
query = """
SELECT
  directors,
  AVG(averageRating) AS media,
  COUNT(originalTitle) AS quantidade_minima
FROM imdb
GROUP BY directors
HAVING quantidade_minima > 5
ORDER BY media DESC
LIMIT 10
"""

result = sqldf(query, locals())
print(result)
```

## ***9 - Diretores que tem os maiores custos de produção médio***

```python
query = """
SELECT
  directors,
  AVG(budget) AS custo,
  AVG(averageRating) AS media
FROM imdb
GROUP BY directors
ORDER BY AVG(budget) DESC
LIMIT 11
"""

result = sqldf(query, locals())
print(result)
```

## ***10 - Anos de produção com maiores notas***

```python
query = """
SELECT
  release_year AS ano,
  SUM(averageRating) as soma
FROM imdb
WHERE ano IS NOT NULL
GROUP BY ano
ORDER BY soma DESC
LIMIT 10
"""

result = sqldf(query, locals())
print(result)
```
